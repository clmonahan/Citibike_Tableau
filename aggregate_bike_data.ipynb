# Steps

1. Drop "Column 1"
2. Clean up headings
3. Combine DFs

```python
# Drop column 1 from q19
Q2_19 = Q2_19.drop(['Column 1'], axis = 1)
```

```python
Q2_19.head()
```

```python
# Append Q2 & Q3 from 2017 into same DF
df_17 = Q2_17.append(Q3_17)
df_17.head()
```

```python
# Append Q2 & Q3 from 2018 into same DF
df_18 = Q2_18.append(Q3_18)
df_18.head()
```

```python
# Append 2017 & 2018 into same DF
Y1718 = df_17.append(df_18)
Y1718.head()
```

```python
# Add 2019 info to new DF
complete_df = Y1718.append(Q2_19)
complete_df.head()
```

```python
# Rename Columns
complete_df.columns = ['Date', 'Total Trips', 'Total Miles', 'Total Annual Members', '24-hour Passes Purchased', '3-day passes purchased']
complete_df.head()
```
