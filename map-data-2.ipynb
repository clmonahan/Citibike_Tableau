```python
# from datetime import datetime, timedelta
import datetime as dt
# Dependencies
import pandas as pd
import numpy as np
```

```python
# Load data
bike_df = pd.read_csv('JC-201905-citibike-tripdata.csv')
bike_df.head()
```

```python
# Create a pre-processed copy
# Create a copy of the current state of the df DataFrame
trip_df = bike_df.copy()
```

```python
trip_df.head()
```

```python
# Check for missing data
trip_df.isnull().any()
```

```python
# Info on df
trip_df.info()
```

```python
# Drop columns we won't be using
trip_df = trip_df.drop(['bikeid'], axis = 1)
```

```python
trip_df.head()
```

```python
# Get a list and count of stations
sorted(trip_df['start station name'].unique())
```

```python
# Number of rides leaving from each station
station_starts = trip_df['start station name'].value_counts()
station_starts
```

```python
# Groupby station name
station_name = trip_df.set_index(['end station name'])['end station id']
station_name
```

```python
# Latitude
latitude = trip_df.groupby(['end station name']).mean()
latitude
```

```python
# Number of rides ending at each station
station_ends = trip_df['end station name'].value_counts()
station_ends
```

```python
# Extract Latitude
station_lat = trip_df.set_index(['start station name'])['start station latitude']
station_lat
```

```python
# Extract Longitude
station_lon = trip_df.set_index(['start station name'])['start station longitude']
station_lon
```

```python
# New DF with rides per station
station_rides_df = pd.DataFrame({
    '# Rides Starting Here': station_starts,
    '# Rides Leaving Here' : station_ends,
#     'Station Latitude' : station_lat,
#     'Station Longitude' : station_lon
})
station_rides_df
```

```python
# Convert NaNs to 0
station_rides_df['# Rides Starting Here'].fillna(0)
station_rides_df
```

```python
# To csv
export_csv = trip_df.to_csv ('stationInfoLarge.csv', index = None, header=True)
```
